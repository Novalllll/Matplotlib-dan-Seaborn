- Import NumPy, Matplotlib dan Seaborn

```python
# Import NumPy
import numpy as np

# Import Matplotlib PyPlot
import matplotlib.pyplot as plt

# Import seaborn sebagai sns
import seaborn as sns
```

- Pada excercise ini, gambarkanlah sebuah Line Plot untuk menunjukkan penambahan kasus positif COVID-19 pada bulan April hingga Agustus di Indonesia, prov. DKI Jakarta, dan prov. Jawa Timur (Data aggregasi dari laporan BNPB).

    1. Ubah style menjadi `ggplot`
    2. Gambarkan Line Plot dengan data `bulan` pada x-axis dan data `kasus_x` pada y-axis
    3. Tambahkan `Grid` pada plot
    4. Ubah tick y-axis dengan menggunakan `np.arange()` agar tick y-axis memunculkan data dari 0 hingga 80000 dengan step tiap 10000
    5. Tampilkan label x-axis dengan text `Bulan`
    6. Tampilkan label y-axis dengan text `Tambahan Kasus positif`
    7. Tampilkan judul dengan text `Penambahan Kasus Positif COVID-19 Bulanan`
    8. Tampilkan legenda di kiri atas plot

- Cek style PyPlot

```python
print(plt.style.available)
```

- Set style menjadi `ggplot`

```python
plt.style.use('ggplot')
```

- List value untuk data di Indonesia, DKI Jakarta dan Jawa Timur

```python
# list value bulan
bulan = ['april', 'mei', 'juni', 'juli', 'agustus']


# list value penambahan kasus di Indonesia, DKI Jakarta dan Jawa timur
kasus_indonesia = [8590, 16355, 29912, 51991, 66420]
kasus_jkt = [3428, 3173, 4076, 9975, 18687]
kasus_jatim = [865, 3899, 7279, 9953, 11454]
```

```python
# Set figure dan axes
fig, ax = plt.subplots()


# Gambarkan line plot
sns.lineplot(y=kasus_indonesia, x=bulan)
sns.lineplot(y=kasus_jkt, x=bulan)
sns.lineplot(y=kasus_jatim, x=bulan)
# Reminder: untuk overlay plot harus menggunakan axes level function

# Tambahkan grid
plt.grid(True)

# Tentukan tick y-axis
plt.yticks(np.arange(0, 80000, 10000))

# Tampilkan label untuk x-axis
plt.xlabel('Bulan')

# Tampilkan label untuk y-axis
plt.ylabel('Tambahan kasus positif')
# Tampilkan judul
plt.title('Penambahan Kasus Positif COVID-19 Bulanan')

# Tampilkan legenda
plt.legend(['Indonesia', 'DKI Jakarta', 'Jawa Timur'])

# Tunjukkan plot yang telah digambar
plt.show()
```

# Excercise 2

- Gunakan data pada Excercise 1 untuk menggambarkan figure dengan 3 subplot:
    1. Subplot 1 menggambarkan data bulanan penambahan kasus di Indonesia
    2. Subplot 2 menggambarkan data bulanan penambahan kasus di DKI Jakarta
    3. Subplot 3 menggambarkan data bulanan penambahan kasus di Jawa Timur

```python
# Buat 1 figure dengan 3 obyek axes menggunakan plt.subplots()
fig, axes = plt.subplots(3, 1, sharey=True, sharex=True, figsize=(10, 15))


# Gambarkan plot pada axes pertama
axes[0].set_title('Indonesia')
#axes[0].set_yticks(np.arange(0, 70000, 10000))
sns.lineplot(ax=axes[0], x=bulan, y=kasus_indonesia)

# Gambarkan plot pada axes kedua
axes[1].set_title('Jakarta')
#axes[1].set_yticks(np.arange(0, 70000, 10000))
sns.lineplot(ax=axes[1], x=bulan, y=kasus_jkt)

# Gambarkan plot pada axes ketiga
axes[2].set_title('Jawa Timur')
#axes[2].set_yticks(np.arange(0, 70000, 10000))
sns.lineplot(ax=axes[2], x=bulan, y=kasus_jatim)
# Tunjukkan plot yang telah digambar
plt.show()
```
